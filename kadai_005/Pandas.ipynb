# 16〜DataFrameのテーブル結合を完全解説(merge, join, concat)〜

```python
import pandas as pd
df1 = pd.DataFrame({ 'Key': ['k0', 'k1', 'k2'],
        'A': ['a0', 'a1', 'a2'],
        'B': ['b0', 'b1', 'b2']})

df2 = pd.DataFrame({ 'Key': ['k0', 'k1', 'k2'],
        'C': ['c0', 'c1', 'c2'],
        'D': ['d0', 'd1', 'd2']})
```

```python
df1.merge(df2)
```

```python
# 縦
pd.concat([df1, df2], axis=0)
```

```python
# 横
pd.concat([df1, df2], axis=1)
```

```python
df1 = pd.DataFrame({ 'Key': ['k0', 'k1', 'k2'],

                    'A': ['a0', 'a1', 'a2'],

                    'B': ['b0', 'b1', 'b2']})



df2 = pd.DataFrame({ 'Key': ['k0', 'k1', 'k3'],

                    'C': ['c0', 'c1', 'c3'],

                    'D': ['d0', 'd1', 'd3']})
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='inner')
```

```python
df1 = pd.DataFrame({ 'Key': ['k0', 'k1', 'k2'],

                    'ID': ['aa', 'bb', 'cc'],

                    'A': ['a0', 'a1', 'a2'],

                    'B': ['b0', 'b1', 'b2']})



df2 = pd.DataFrame({ 'Key': ['k0', 'k1', 'k3'],

                    'ID': ['aa', 'bb', 'cc'],

                    'C': ['c0', 'c1', 'c3'],

                    'D': ['d0', 'd1', 'd3']})
```

```python
# KeyカラムをKeyにして結合

df1.merge(df2, on='Key')
```

```python
# IDカラムをKeyにして結合

df1.merge(df2, on='ID')
```

```python
# suffixesを指定する

df1.merge(df2, on='ID', suffixes=('_left', '_right'))
```

```python
df1 = pd.DataFrame({ 'Key1': ['k0', 'k1', 'k2'],

                    'A': ['a0', 'a1', 'a2'],

                    'B': ['b0', 'b1', 'b2']})



df2 = pd.DataFrame({ 'Key2': ['k0', 'k1', 'k3'],

                    'C': ['c0', 'c1', 'c3'],

                    'D': ['d0', 'd1', 'd3']})
```

```python
# suffixesを指定する

df1.merge(df2, left_on='Key1', right_on='Key2')
```

```python
df1.merge(df2, left_index=True, right_index=True)
```
